```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Cleaning Up Data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

```

```python
df.index
```

```python
df.index.name = 'job_index'
df.head(5)
```

```python
df.index.dtype
```

```python
df_Spain = df[df['job_country'] == 'Spain']
```

```python
df_Spain.head(5)
```

```python
df_Spain.reset_index(inplace=True)
df_Spain
```

```python
df_Spain.set_index('job_index', inplace=False)
df_Spain
```

```python
median_pivot = df_Spain.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
median_pivot.sort_index(inplace=True)
```
