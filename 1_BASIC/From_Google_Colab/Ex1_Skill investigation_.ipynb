Filter what positions to apply for based on a list of skills.

```python
exercise_positions = [
    {"role": "Exercise Physiologist", "skills": ["Conduct exercise testing", "Design fitness programs", "Analyze physical health metrics"]},
    {"role": "Clinical Exercise Physiologist", "skills": ["Monitor patients during exercise", "Assess risk factors", "Create personalized exercise plans"]},
    {"role": "Physiotherapist", "skills": ["Develop rehabilitation plans", "Provide physical therapy exercises", "Guide injury prevention strategies"]},
    {"role": "Personal Trainer", "skills": ["Create customized fitness plans", "Lead one-on-one sessions", "Track client progress"]},
    {"role": "Exercise Physiologist", "skills": ["Perform fitness assessments", "Lead group exercise sessions", "Offer nutrition advice"]},
    {"role": "Athletic Trainer", "skills": ["Provide injury prevention", "Manage rehabilitation for athletes", "Assess recovery progress"]},
    {"role": "Fitness Instructor", "skills": ["Lead group fitness classes", "Offer exercise technique guidance", "Motivate participants"]},
    {"role": "Cardiac Rehabilitation Specialist", "skills": ["Supervise cardiovascular recovery", "Monitor vitals", "Develop heart-healthy exercise plans"]},
    {"role": "Wellness Coach", "skills": ["Conduct physical activity counseling", "Promote healthy lifestyle habits", "Assist with goal setting"]},
    {"role": "Strength and Conditioning Coach", "skills": ["Enhance athletic performance", "Design strength programs", "Educate on injury prevention"]}
]
```

```python
my_skills = ["Conduct exercise testing", "Design fitness programs"]
```

```python
qualified_roles = []

for position in exercise_positions:
  qualified = True

  for skill in my_skills:
    if skill not in position['skills']:
      qualified = False
      break
  if qualified:
    qualified_roles.append(position['role'])

print(qualified_roles)
```

```python
class Coordinate:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    x_dist = (self.x - other.x) ** 2
    y_dist = (self.y - other.y) ** 2
    return (x_dist + y_dist) ** 0.5

c1 = Coordinate(3, 4)
c2 = Coordinate(4, 5)

print(c1)
print(c2)
```
