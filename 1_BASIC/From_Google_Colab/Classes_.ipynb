```python
class DanielList:
  def __init__(self):
    """Inizializes an empty list."""
    self._items = []

  def add(self, item):
    """Adds a new item to the end of the list."""
    self._items.append(item)

  def __getitem__(self, index):
    """Retrieves and item by index."""
    return self._items[index]

  def __setitem__(self, index, value):
    """Sets an item at a specific index"""
    self._items[index] = value

  def __repr__(self):
    """Returns a string representation of the list."""
    return str(self._items)

  def __len__(self):
    """Returns the number of items in the list."""
    return len(self._items)
```

```python
def calculate_salary(base_salary, bonus_rate=.1):
  """
  Calculate the total salary based on the base salary and bonus rate.

  Args:
  base_salary (float): The base salary.
  bonus_rate (float, optional): The bonus rate. Defaults to 0.1.

  Returns:
  float: The total salary.
  """
  return base_salary * (1 + bonus_rate)

  def calculate_bonus(total_salary, base_salary):
    """
    Calculate the bonus based on the total salary and base salary.

    Args:
    total_salary (float): The total salary.
    base_salary (float): The base salary.

    Returns:
    float: The bonus.
    """
    return (total_salary - base_salary) / base_salary


```

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=.1, symbol="$"):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total_salary = self.base_salary * (1 + self.bonus_rate)
    self.bonus = self.total_salary - self.base_salary

  def __repr__(self):
    return f"{self.symbol}{self.base_salary:,.0f}"

  def calculate_salary(self):
    """
    Calculate the total salary based on the base salary and bonus rate.

    Args:
    base_salary (float): The base salary.
    bonus_rate (float, optional): The bonus rate. Defaults to 0.1.

    Returns:
    float: The total salary.
    """
    return f"{self.symbol}{self.total_salary:,.0f}"

  def calculate_bonus(self):
    """
    Calculate the bonus based on the total salary and base salary.

    Args:
    total_salary (float): The total salary.
    base_salary (float): The base salary.

    Returns:
    float: The bonus.
    """
    return f"{self.symbol}{self.bonus:,.0f}"
```
