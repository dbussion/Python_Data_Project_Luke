```python
from datasets import load_dataset
```

```python
import pandas as pd
```

```python
dataset = load_dataset('lukebarousse/data_jobs')
```

```python
df = dataset['train'].to_pandas()
```

```python
df['job_country'].value_counts().head(20)
```

```python
spain_jobs = df[df['job_country'] == 'Spain']
```

```python
spain_jobs = spain_jobs[spain_jobs['salary_year_avg'].notna()]
spain_jobs
```

```python
spain_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)
```
