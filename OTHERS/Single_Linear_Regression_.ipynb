```python
# Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load Dataset
df = pd.read_csv("https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv")

```

```python
df
```

```python
# Select Variables
x = df[['housing_median_age']]
y = df[['median_house_value']]
```

```python
# Split into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Create and train model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Predict
y_pred = model.predict(x_test)
```

```python
# Evaluate
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
print(f'Model Coefficient: {model.coef_.item():.2f}')
print(f'Intercept: {model.intercept_.item():.2f}')
```

```python
# Scatter plot with regression line
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_test, y_pred, color='red', label='Predicted')
plt.xlabel('Housing median age (years)')
plt.ylabel('Housing median value ($)')
plt.title('Simple Linear Regression - Housing')
plt.legend()
plt.show()
```
